```python
df_sample = pd.read_csv('transactions_dataset.csv', nrows=2e6, sep=';')
```

```python
df_sample.describe()
```

```python
# Data types and missing values
data_types = df_sample.dtypes
missing_values = df_sample.isnull().sum()

print("Data Types:\n", data_types)
print("\nMissing Values:\n", missing_values)
```

```python
# Distribution of sales_net
plt.figure(figsize=(10, 5))
plt.hist(df_sample['sales_net'], bins=50, edgecolor='k')
plt.title('Distribution of Sales Net')
plt.xlabel('Sales Net')
plt.ylabel('Frequency')
plt.show()

# Distribution of quantity
plt.figure(figsize=(10, 5))
plt.hist(df_sample['quantity'], bins=50, edgecolor='k')
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of order_channel
order_channel_distribution = df_sample['order_channel'].value_counts()
print(order_channel_distribution)

# Plot the distribution
plt.figure(figsize=(10, 5))
order_channel_distribution.plot(kind='bar')
plt.title('Distribution of Order Channel')
plt.xlabel('Order Channel')
plt.ylabel('Frequency')
plt.show()
```
